#### Python Crash Course

In this tutorial we will cover some basic aspects of Python using IPython (Jupyter) notebooks.

- Syntax
- Data types
- Loops and control structures
- numpy, scipy, math

Operation:

```python
# opreation: 
2 + 2 # addition
5 - 3 # subtraction
4 * 2 # multiplication
8 / 4 # division
9 % 4 # modulus
2 ** 3 # exponentiation 
10 // 3 # floor division
```

Comparison:

```python
# comparison:
5 > 3 # greater than
2 < 4 # less than
3 >= 3 # greater than or equal to
2 <= 5 # less than or equal to
3 == 3 # equal to
4 != 2 # not equal to
```

Logical operation:

```python
# logical:
True and False # logical AND
True or False # logical OR
not True # logical NOT
```

bitwise operation : a operation that works on numbers at the level of their individual bits, comparing and manipulating them using logical operators like AND, OR, XOR, and NOT

```python
# bitwise:
5 & 3 # bitwise AND 
5 | 3 # bitwise OR
5 ^ 3 # bitwise XOR
```

```python
# shift:
4 << 1 # left shift
8 >> 2 # right shift
```

Datatype:

```python
# datatype:
type(5) # integer
type(3.14) # float
type("Hello") # string
type(True) # boolean
type([1, 2, 3]) # list
type((1, 2, 3)) # tuple 
type({ "a": 1, "b": 2}) # dictionary
type({1, 2, 3}) # set
# NoneType
type(None)
# complex
type(2 + 3j)
# byte
type(b'Hello')
# bytearray
type(bytearray(b'Hello'))
# range
type(range(5))
```

assignment values:

```python
# assignment:
x = 5 # simple assignment
x += 2 # addition assignment
x -= 1 # subtraction assignment
x *= 3 # multiplication assignment
x /= 2 # division assignment
x %= 4 # modulus assignment
x **= 2 # exponentiation assignment
x //= 3 # floor division assignment
```

```python

# identity:
a = [1, 2, 3]
b = a
a is b # identity operator
a is not b # not identity operator
# membership:
lst = [1, 2, 3, 4]
2 in lst # membership operator
5 not in lst # not membership operator
```

control flow:

```python
# if elsif else:
if x > 0:
    print("x is positive")
elif x == 0:
    print("x is zero")
else:
    print("x is negative")
    
# ternary:
x = 10  
"High" if x > 5 else "Low" # ternary operator   
```

```python
#looping:

for i in range(5):
    print(i)

while x > 0:
    print(x)
    x -= 1
```

Function:

```python
# difine function:
def add(a, b):
    return a + b
## calling the function
result = add(3, 5)

# lambda:
square = lambda x: x ** 2 # lambda function
square(4) # calling lambda function
```

define class:

```python
# attribute access:
class MyClass:
    def __init__(self):
        self.value = 10
obj = MyClass()
obj.value # attribute access
```

Difference between List & Tuple & Set & Dict:

```python
# list comprehension:
squared = [x ** 2 for x in range(5)] # list comprehension
# dictionary comprehension:
squared_dict = {x: x ** 2 for x in range(5)} # dictionary comprehension
# set comprehension:
squared_set = {x ** 2 for x in range(5)} # set comprehension
# generator expression:
squared_gen = (x ** 2 for x in range(5)) # generator expression
```

```python
# indexing:
lst = [10, 20, 30, 40]
lst[2] # indexing
# slicing:
lst[1:3] # slicing
```
