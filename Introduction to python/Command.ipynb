#### Overview about conda famous conda commands

Note: these commands execute at terminal where's conda defined.

To display a list of commands at conda:  ```conda``` at terminal

To create environment: ``` conda create -n "MyEnvironmentName" python==3.XX ```

To activate the environment at terminal: ``` conda activate "MyEnvironmentName" ```

To deactivate the environment at terminal: ``` conda deactivate ```

Display a list of conda environment at your PC: ``` conda env list ```

To remove environment: ``` Conda remove -n "MyenvironmentName" ```

To export and sharing environments: ``` conda env export > myenvironment.yml ```
